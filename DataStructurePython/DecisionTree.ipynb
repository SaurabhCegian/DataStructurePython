## ***`Decision Tree`***

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
data = pd.read_csv("/content/data_cleaned.csv")
data.head()
```

```
data.shape
```

```
#checking missing value in data
data.isnull().sum()
```

```
# Separating dependent and independent variable
y = data["Survived"]
X = data.drop(["Survived"], axis = 1)
```

```
# Validation using train_test_split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=101, stratify=y, test_size=0.25)
```

```
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import DecisionTreeRegressor
```

```
dt_model = DecisionTreeClassifier(random_state=10)
```

```
dt_model.fit(X_train, y_train)
```

```
# Checking the training score
dt_model.score(X_train, y_train)
```

```
# Checking the testing score
dt_model.score(X_test, y_test)
```

```
dt_model.predict(X_test)
```

```
dt_model.predict_proba(X_test)
```

```
y_pred = dt_model.predict_proba(X_test)[:,1]
```

```
new_y = []
for i in range(len(y_pred)):
  if(y_pred[i] < 0.6):
    new_y.append(0)
  else:
    new_y.append(1)
```

```
from sklearn.metrics import accuracy_score
accuracy_score(y_test, new_y)
```
