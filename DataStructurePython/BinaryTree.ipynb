```
class Node:
  def __init__(self, data):
    self.data = data
    self.left = None
    self.right = None
  def insert(self, root, data):
    queue = []
    queue.append(root)
    while(len(queue)):
      temp = queue[0]
      queue.pop(0)
      if( not temp.left):
        temp.left = Node(data)
        break
      else:
        queue.append(temp.left)
      
      if(not temp.right):
        temp.right = Node(data)
        break
      else:
        queue.append(temp.right)
   
  def levelorder(self, root):
    if(root == None):
      return
    queue = []
    queue.append(root)
    while(len(queue)):
      temp = queue[0]
      print(temp.data, end=" ")
      queue.pop(0)
      if(temp.left):
        queue.append(temp.left)
      if(temp.right):
        queue.append(temp.right)
        
  def inorder(self, root ):
    if(root == None):
      return
    current  = root
    stack = []
    while(True):
      if(current is not None):
        stack.append(current)
        current = current.left
      elif(stack):
        current = stack.pop()
        print(current.data, end = " ")
        current = current.right
      else:
        break
        
  def preorder(self, root):
    if(root == None):
      return
    stack = []
    stack.append(root)
    while(len(stack)):
      temp = stack.pop()
      print(temp.data, end=" ")
      if(temp.right):
        stack.append(temp.right)
      if(temp.left):
        stack.append(temp.left)
   
  def height_of_tree(self, root):
    if(root == None):
      return 0
    q = []
    q.append(root)
    height = 0
    while(True):
      node_count = len(q)
      if(node_count == 0):
        print(height)
        return height
      height += 1
      while(node_count > 0):
        temp = q[0]
        q.pop(0)
        if(temp.left):
          q.append(temp.left)
        if(temp.right):
          q.append(temp.left)
        node_count -= 1
        
  def rightmostnode(self, root):
    if(root == None):
      return None
    stack = []
    current = root
    while(current != None):
      stack.append(current)
      current = current.right
    return stack.pop().data
  
  def delete_node(self, root, key):
    if(root == None):
      return
    queue = []
    queue.append(root)
    while(len(queue)):
      temp = queue[0]
      if(temp.data == key):
        temp.data = self.rightmostnode(root)
      queue.pop(0)
      if(temp.left):
        queue.append(temp.left)
      if(temp.right):
        queue.append(temp.right)
        
  def level_spiral_order(self, root):
    if(root == None):
      return
    stack1 = []
    stack2 = []
    stack1.append(root)
    while(stack1 or stack2):
      while(stack1):
        temp = stack1.pop()
        print(temp.data, end=" ")
        if(temp.right):
          stack2.append(temp.right)
        if(temp.left):
          stack2.append(temp.left)
      while(stack2):
        temp = stack2.pop()
        print(temp.data, end=" ")
        if(temp.left):
          stack1.append(temp.left)
        if(temp.right):
          stack1.append(temp.right)
          
  def nth_of_inorder(self, root, n):
    if(root == None):
      return
    stack = []
    current = root
    count = 0
    while(True):
      if(current is not None):
        stack.append(current)
        current = current.left
      elif(stack):
        current = stack.pop()
        count += 1
        if(count == n):
          print(current.data,end=" ")
          break
        current = current.right
      else:
        break
  def printpostorder(self, inorder, preorder):
    print(inorder)
    i = 1
    while(i < inorder.index(preorder[0])):
      temp = inorder[i]
      inorder[i] = inorder[i+1]
      inorder[i+1] = temp
      i = i+2
    index = inorder.index(preorder[0]) + 2
    print(index)

            
root = Node(1)
root.insert(root, 2)
root.insert(root, 3)
root.insert(root, 4)
root.insert(root, 5)
root.insert(root, 6)
# root.insert(root, 7)
# root.insert(root, 8)
# root.insert(root, 9)

# root.levelorder(root)
# print()
# root.inorder(root)
# print()
# root.preorder(root)
# print()
# root.height_of_tree(root)
# print()
# root.delete_node(root, 3) 
# print()
# root.levelorder(root)
# print()
# root.level_spiral_order(root)
# print()
# root.nth_of_inorder(root, 4)
# print()
```
